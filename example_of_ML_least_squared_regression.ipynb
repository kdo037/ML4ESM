```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1, 100, 100)
```

```python
y = x + 500
```

```python
noise = np.random.normal(0, 10, y.shape)
```

```python
y_noise = np.subtract(y, noise)
```

```python
plt.scatter(x, y_noise)
plt.xlabel('x')
plt.ylabel('y')
```

```python
# initial theta0 = 0.0 1and theta1 = 0.01
b = 0.1
w = 0.1

# learning rate alpha = 0.01
a = 0.000001
```

```python
for i in range(1, 200000):
    # compute predicted y_hat = theta0 + theta1*x
    y_hat = b + w*x

    # compute the loss function for theta0
    l0 = np.sum(y_hat-y)
    l1 = np.sum(np.dot((y_hat-y_noise), x))

    b = b - a*l0
    w = w - a*l1
print('y_hat = ', round(b, 2), ' + ', round(w, 2), 'x')
new_y = b + w*x
plt.scatter(x, y_noise)
plt.plot(x, new_y, 'red')
print('Loss: ' + str(l0)[0:6])
```
